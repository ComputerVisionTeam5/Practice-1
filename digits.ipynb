```python
thr = 60
image = cv.imread('digits/3.jpeg')
img = crop_image(image) 
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

homo_img = homomorphic_filter(gray_img)

_, binary_img = cv.threshold(homo_img, thr, 255, cv.THRESH_BINARY)

binary_img = -binary_img + 255

kernel = np.ones((5,5),np.uint8)
#opened_img = cv.morphologyEx(binary_img, cv.MORPH_OPEN, kernel)

dilated_img = cv.dilate(binary_img,kernel,iterations = 1)

plot_img(4, (12, 6), ["Gray", "Homomorphic", "Binary", "Dilated"], [gray_img, homo_img, binary_img, dilated_img])
```
